```python
### 1. Setup envoronment

import numpy as np
import pandas as pd
import pandas_datareader as pdr
from pandas.plotting import register_matplotlib_converters
register_matplotlib_converters()
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
### 2. Get Data
stocks = "AMZN GOOG NFLX FB MSFT GLD".split()
stocks
```

```python
data = pdr.get_data_yahoo(stocks, '20200101')['Close']
```

```python
data.tail()
```

```python
### 3. Basic descriptive statistics
```

```python
### mean() function = avarage price diring the selected time frame
data.mean()
```

```python
### quantile() function = return the 25% tail on the selected data on a given time frame
data.quantile(.25)
```

```python
### describe() function = retrieves all the statistics operations
data.describe()
```

```python
### 4. Typical Visualizations
```

```python
plt.plot(data)
```

```python
## To see one of the data points plotted, or one of the columns
plt.plot(data['AMZN'])
```

```python
## To present how is this stock (price) change Tumbs UP!
porcent_change = np.log(data).diff()
porcent_change.head()
```

```python
## To delete the first column which is ther NaN row, since we can't calculate for the first one
porcent_change.dropna(inplace=True)
```

```python
change = data.pct_change()
change
```

```python
amazon_stock = porcent_change['AMZN']
amazon_stock.describe()
```

```python
### To have an idea of the shape, we can unse the histogram
plt.hist(amazon_stock, bins=35, color='g', edgecolor='w');
```

```python
### Now is worthy to mention we can use seaborn as well, for the same porpuse
sns.distplot(porcent_change['AMZN'], kde_kws={'label':'AMZN'});
```

```python
### Seaborn also allow us to compare it with another group (stock). Google vs Amazon
sns.distplot(porcent_change['AMZN'], kde_kws={'label':'AMZN'});
sns.distplot(porcent_change['GOOG'], kde_kws={'label':'GOOG'});
```

```python
### How to know how related those prices are form each other
sns.scatterplot(porcent_change['AMZN'], porcent_change['GOOG']);
```

```python
### If we want to know exactly how much is the correlation is
porcent_change.corr()
```

```python
### Another way to look this correlation matrix is through a scatter plot
sns.pairplot(porcent_change)
```
